```python
import cv2
from google.colab.patches import cv2_imshow
```

Original

```python
img=cv2.imread('HORSE.jpg',1)
#cv2.imshow("Seven Horses",img)
cv2.imwrite("horse.png",img)
cv2_imshow(img)
```

Black and white

```python
obj=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imwrite("grayt_horses.png",obj)
cv2_imshow(obj)
```

Grayscale

```python
thre,dat=cv2.threshold(obj,115,245,cv2.THRESH_BINARY)
cv2.imwrite("gs.png",dat)
cv2_imshow(dat)
```

```python
hrs_img=img[:,:,0]
cv2.imwrite("hrs_img_edit.jpeg",hrs_img)
cv2_imshow(hrs_img)
```

Cropped Image

```python
crp_img=img[150:250,300:450]
cv2_imshow(crp_img)
```

Threshold Image

```python
tm,trs=cv2.threshold(img,115,245,cv2.THRESH_BINARY)
cv2_imshow(trs)
```
